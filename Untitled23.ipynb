# F.Convert non-empty ‘avg pace’ into a float that is still representative of the information contained in the column.

```python
for i in range(len(df)):
    if not pd.isna(df['avg_pace'][i]):
        df['avg_pace'][i] = float(df['avg_pace'][i].split(sep=':')[1])/60 + int(df['avg_pace'][i].split(sep=':')[0])
    
df['avg_pace'] = df['avg_pace'].astype(float)
```
